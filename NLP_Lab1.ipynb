```python
#Function for expanding all the contractions in the paragraph
c_re = re.compile('(%s)' % '|'.join(Contractions.keys()))

def expandContractions(text, c_re=c_re):
    def replace(match):
        return Contractions[match.group(0)]
    return c_re.sub(replace, text.lower())
```

   ### Removing Punctuations from the string 

```python
def Punctuation(string): 
  
    # punctuation marks 
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
  
    # traverse the given string and if any punctuation 
    # marks occur replace it with null 
    for x in string: 
        if x in punctuations: 
            string = string.replace(x, "") 
    return string
```

```python
Expanded_para=expandContractions(firstpara)
Filtered_para=Punctuation(Expanded_para)
```

### Tokenization

```python
from nltk.tokenize import word_tokenize
tokens=word_tokenize(Filtered_para)
tokens
```

### Removing Stopwords

```python
#Removing stopwords using NLTK coprus library
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
words = [w for w in tokens if not w in stop_words]
words
```

### Stemming

```python
#Stemming using using NLTK's porterStemmer class
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
stemmed = [porter.stem(word) for word in words]
stemmed
```

### Lemmatization

```python
#Lemmatization using NLTK's porterStemmer class
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 
lemmatized=[lemmatizer.lemmatize(word) for word in words]
lemmatized
```

```python
no_of_same_words=0
total_words=0
for i,j in zip(stemmed,lemmatized):
    if i==j:
        no_of_same_words+=1
    total_words+=1
print("Number of words giving same stem in both methods are:",no_of_same_words)
print("Total number of words are:",total_words)
```
